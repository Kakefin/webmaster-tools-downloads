```python
#array of all head terms
head_terms = [ 
              'cloud computing',
              'cloud server',
              'cloud services',
              'cloud products']
#array of arrays, each representing a product and category
product_category_pairs = [
                    ['dynamodb', 'database'],
                    ['elasticache', 'caching'],
                    ['athena', 'analytics'],
                    ['s3', 'storage'],
]
#suffix modifiers
suffix_modifiers = [
             'pricing',
             'cost',
             'tutorial',
             'training',
             'customers',
             'use cases',
             'features',
             'benefits',
             'case studies',
             'faqs',
             'support'
             ]

#prefix modifiers
prefix_modifiers = [
                    'best',
                    'scalable',
                    'extensible',
                    'what is',
                    'how to',
                    'free',
]
#loop through all head terms
for product_category_pair in product_category_pairs:
    #split up city and state pairs
    product = product_category_pair[0]
    category = product_category_pair[1]

    for head_term in head_terms:

      #print combinations without modifiers
      print("Brand: " + product)
      print("Non-Brand - Head: " +  category)
      print("Non-Brand - Torso: " +  head_term + " " + category)

      #print combinations without modifiers
      for suffix_modifier in suffix_modifiers:
        print("Non-Brand - Long-tail: " + head_term + " " + category + " " + suffix_modifier)

      for prefix_modifier in prefix_modifiers:
        print("Non-Brand - Long-tail: " + prefix_modifier + " " + head_term + " " + category)
      print(' ')

```
